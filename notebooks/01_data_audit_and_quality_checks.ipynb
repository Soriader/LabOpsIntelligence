```python
import pandas as pd
import numpy as np
import DataInspector as di

# ========================================================================================================================
# Load raw data
# ========================================================================================================================

df_lab_events = pd.read_csv("../data/raw/lab_events.csv", encoding="utf-8")
df_samples = pd.read_csv("../data/raw/samples.csv", encoding="utf-8")
df_tests = pd.read_csv("../data/raw/tests.csv", encoding="utf-8")

# ========================================================================================================================
# Define validation rules
# ========================================================================================================================

UNIT_RULES = {
    "sulfur": "mg/kg",
    "water": "mg/kg",
    "density": "kg/m3",
    "viscosity": "cSt",
    "chloride": "mg/kg",
    "flash_point": "C",
}

ALLOWED_PRIORITY = ["normal", "high", "urgent"]
ALLOWED_STATUS = ["OK", "ERROR", "RETEST"]

# ========================================================================================================================
# Audit: samples
# ========================================================================================================================

di.DataInspector.display_dimensions(df_samples, "df_samples")

di.DataInspector.display_unique_value(df_samples, "df_samples", "sample_id")
di.DataInspector.display_unique_value(df_samples, "df_samples", "priority")
di.DataInspector.display_unique_value(df_samples, "df_samples", "client_name")

di.DataInspector.display_correct_value(
    df_samples,
    "df_samples",
    "priority",
    ALLOWED_PRIORITY
)

di.DataInspector.display_null_value(df_samples, "df_samples", "client_name")
di.DataInspector.display_null_value(df_samples, "df_samples", "priority")

# ========================================================================================================================
# Audit: tests - basic structure and missing values
# ========================================================================================================================

di.DataInspector.display_dimensions(df_tests, "df_tests")

di.DataInspector.display_unique_value(df_tests, "df_tests", "parameter")
di.DataInspector.display_unique_value(df_tests, "df_tests", "unit")
di.DataInspector.display_unique_value(df_tests, "df_tests", "status")

di.DataInspector.display_null_value(df_tests, "df_tests", "result_value")
di.DataInspector.display_null_value(df_tests, "df_tests", "unit")
di.DataInspector.display_null_value(df_tests, "df_tests", "analyst")

# ========================================================================================================================
# Audit: tests - validation rules
# ========================================================================================================================

di.DataInspector.display_correct_value(
    df_tests,
    "df_tests",
    "status",
    ALLOWED_STATUS
)

di.DataInspector.units_correctness(
    df_tests,
    "df_tests",
    "parameter",
    "unit",
    UNIT_RULES
)

di.DataInspector.unlogical_combination(
    df_tests,
    "df_tests",
    "status",
    "result_value"
)

di.DataInspector.duplicated_in_tests(
    df_tests,
    "df_tests",
    "sample_id",
    "parameter",
    "test_date"
)

di.DataInspector.find_outliers(
    df_tests,
    "df_tests",
    "parameter",
    "result_value",
    "unit"
)

# ========================================================================================================================
# Audit: lab_events - required events and event order
# ========================================================================================================================

di.DataInspector.display_dimensions(df_lab_events, "df_lab_events")

di.DataInspector.display_unique_value(df_lab_events, "df_lab_events", "event_type")

di.DataInspector.validate_required_events_per_sample(
    df_lab_events,
    "df_lab_events",
    "sample_id",
    "event_type"
)

di.DataInspector.check_event_order(
    df_lab_events,
    "df_lab_events",
    "sample_id",
    "event_type",
    "event_timestamp"
)

# ========================================================================================================================
# Audit: relationship samples vs lab_events
# date_received ≈ received event
# ========================================================================================================================

di.DataInspector.check_samples_lab_events_date(
    df_samples,
    df_lab_events,
    "date_received vs received event",
    "sample_id",
    "date_received",
    "event_type",
    "event_timestamp"
)

# ========================================================================================================================
# Audit: relationship tests vs lab_events
# testing_started <= test_date <= testing_finished
# ========================================================================================================================

di.DataInspector.check_tests_vs_events(
    df_tests,
    df_lab_events,
    "tests vs lab_events",
    "sample_id",
    "test_date",
    "event_type",
    "event_timestamp"
)
```
